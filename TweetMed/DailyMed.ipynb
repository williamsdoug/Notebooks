```python
def getContentsXML(name, include_attr=True):
    with open(spl_sample) as fd:
        obj = xmltodict.parse(fd.read(), xml_attribs=include_attr)
        return dict(obj['document'])
```

```python
result = getContentsXML(spl_sample)
```
